```python
# Bibliotecas usadas
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_telco_customer = pd.read_csv('https://raw.githubusercontent.com/SilvioCSOliveira/ChallengeDataScience2022AluraVoz/main/Semana2/df_telco_customer.csv',sep=';')
df_telco_customer.info()
```

```python
df_telco_customer['ClienteCancelado'].value_counts()
```

Analisar a variável target: Churn

```python
# Usar uma pizza ou barplot seaborn para mostrar a proporção de churn
count = df_telco_customer["ClienteCancelado"].value_counts().values
labels = df_telco_customer["ClienteCancelado"].value_counts().index

plt.pie(count,labels = list(labels),autopct='%.0f%%' , textprops = {"fontsize": 12},)  #,colors =["#6a8ee8", "#424ad1"]



plt.title('Percentual de cancelamentos')
plt.legend()
plt.show()
```

```python
# dividir os cancelados e não cancelados para verificar se há alguma grande diferença entre os serviços, e demais caracteristicas
df_telco_customer_churn = df_telco_customer.query('ClienteCancelado == "Yes"')
df_telco_customer_nao_churn = df_telco_customer.query('ClienteCancelado == "No"')

df_telco_customer_nao_churn.info()
```

```python
df_telco_customer_churn["Genero"].value_counts()
```

```python
df_telco_customer_nao_churn["Genero"].value_counts()
```

```python
count_c = df_telco_customer_churn["Genero"].value_counts().values
labels_c = df_telco_customer_churn["Genero"].value_counts().index

count_nc = df_telco_customer_nao_churn["Genero"].value_counts().values
labels_nc = df_telco_customer_nao_churn["Genero"].value_counts().index

plt.figure(figsize = (14, 6))
plt.subplot(1, 2, 1)

plt.pie(count_c,labels = list(labels_c),autopct='%.0f%%' , textprops = {"fontsize": 12},)  #,colors =["#6a8ee8", "#424ad1"]

#plt.pie(df_telco_customer['ClienteCancelado'].value_counts(),autopct='%.0f%%')

plt.title('Cancelados')
plt.legend()

plt.subplot(1, 2, 2)

plt.pie(count_nc,labels = list(labels_nc),autopct='%.0f%%' , textprops = {"fontsize": 12},)  #,colors =["#6a8ee8", "#424ad1"]

#plt.pie(df_telco_customer['ClienteCancelado'].value_counts(),autopct='%.0f%%')

plt.title('Não Cancelados')
plt.legend()


#plt.show()
```
